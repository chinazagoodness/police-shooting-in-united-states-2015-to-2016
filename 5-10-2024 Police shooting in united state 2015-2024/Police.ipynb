# bar chart
A bar chart (or bar graph) is a type of data visualization that uses rectangular bars to represent data values.
Each bar's length or height is proportional to the value it represents, allowing for easy comparison betweencategories.
Key Features of a Bar Chart:
Bars: Represent individual categories (e.g., countries) and their values (e.g., population).

Axes:
The horizontal axis (x-axis) typically represents categories (e.g., country names).

The vertical axis (y-axis) represents numerical values (e.g., population size).

Bar Height/Length: Corresponds to the value of the category (the taller or longer the bar, the higherthevalue).

```python
# creating bars 
import matplotlib.pyplot as plt
# sample data: country names ans their populations 
countries = ['China','India','Usa','Indonesia', 'Pakistan']
population = [1444216107,1393409038,332915073,276361783,225199937]


plt.figure(figsize=(10,6))
plt.bar(countries,population)
```

```python
# to change colour
countries = ['China','India','Usa','Indonesia', 'Pakistan']
population = [1444216107,1393409038,332915073,276361783,225199937]


plt.figure(figsize=(10,6))
plt.bar(countries,population, color = 'green')
```

```python
# adding title and labels
 
countries = ['China','India','Usa','Indonesia', 'Pakistan']
population = [1444216107,1393409038,332915073,276361783,225199937]


bars = plt.figure(figsize=(10,6))
plt.bar(countries,population, color = 'purple')
plt.title('populations of countries', fontsize=16)
```

```python
# to label x and y axis 
countries = ['China','India','Usa','Indonesia', 'Pakistan']
population = [1444216107,1393409038,332915073,276361783,225199937]


plt.figure(figsize=(10,6))
plt.bar(countries,population, color = 'orange')
plt.title('populations of countries', fontsize=16)
plt.xlabel('countries')
plt.ylabel('population')
plt.show()
```

```python
# add label on top of each bar


countries = ['China','India','Usa','Indonesia', 'Pakistan']
population = [1444216107,1393409038,332915073,276361783,225199937]


plt.figure(figsize=(10,6))
bars = plt.bar(countries,population, color = 'red')
plt.title('populations of countries', fontsize=16)
plt.xlabel('countries')
plt.ylabel('population')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2., height*1.01, #adjust position
             f'{height:,}', ha='center', va='bottom', fontsize=10)
# Explanation 
# bars: Holds the bar objects created by plt.bar().

# plt.text(): Adds text labels on top of each bar:

# bar.get_x() + bar.get_width()/2. centers the label on the bar.

 # height * 1.01 positions the label slightly above the bar.

# f'{height:,}' formats the population with commas for better readability.

# ha='center' and va='bottom': Align the text horizontally in the center and vertically at the bottomofthetext.
```

```python
# horizontal bar chart 
countries = ['China','India','Usa','Indonesia', 'Pakistan']
population = [1444216107,1393409038,332915073,276361783,225199937]

# creating horizontal bar
plt.figure(figsize=(10,6))
bars = plt.barh(countries,population, color = 'skyblue')
```

```python
 #Location-Based Analysis
Which states have the highest number of policeshootings?
state_counts = police['state'].value_counts()

state_counts
```

```python
# Ensure the 'date' column is in datetime format
police['date'] = pd.to_datetime(police['date'], errors='coerce')
```

```python
#Time-Based Analysis
# In which months do most police shootingsoccur?

incident_per_month = police['date'].dt.month.value_counts()

incident_per_month
```

```python
# What is the most common race of the people involved in the policeshootings?
most_common_race = police['race'].value_counts()
gender_counts = police['gender'].value_counts()
most_common_race,gender_counts
```

```python
#How many incidents occurred eachyear?
incidents_per_year = police['date'].dt.year.value_counts()

incidents_per_year
```

```python
# Weapon Involvement
# What proportion of incidents involve individuals carryingaweapon?
# Assuming there's a 'weapon' or similar column
weapon_involvement = police['armed'].value_counts(normalize=True)

weapon_involvement

# NB:normalize=True: This option returns the relative frequencies (or proportions) instead of the absolute counts. 
#Instead of showing how many times each value appears, it shows what percentage of the total rows that valuerepresents
```
